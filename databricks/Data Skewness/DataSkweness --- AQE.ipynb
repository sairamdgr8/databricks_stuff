```
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
spark_dis_AQE=SparkSession.builder.appName("Data_skewness_before_enabling_AEQ").getOrCreate()

spark_dis_AQE.conf.set("spark.sql.adaptive.enabled", "false")
spark_dis_AQE.conf.set("spark.sql.autoBroadcastJoinThreshold", "-1")
spark_dis_AQE.conf.set("spark.sql.shuffle.partitions", "3")


```

```
df1 = spark_dis_AQE.range(1000000000) \
    .withColumn("id", lit("x"))

extravalues = spark_dis_AQE.range(4)\
    .withColumn("id", lit("y"))

moreextravalues = spark_dis_AQE.range(4)\
    .withColumn("id", lit("z"))

df1 = df1.union(extravalues).union(moreextravalues)
df1.display(10)
df1.select('id').distinct().collect()
```

```
df2=spark_dis_AQE.range(1000000000) \
    .withColumn("id", lit("x"))
extravalues1 = spark_dis_AQE.range(1)\
    .withColumn("id", lit("y"))

moreextravalues1 = spark_dis_AQE.range(1)\
    .withColumn("id", lit("z"))

df2 = df2.union(extravalues1).union(moreextravalues1)
df2.display(10)
```

```
## join 

df1.join(df2,'id').select("id").groupBy("id")\
    .agg(count("*"))\
    .show(100, False)\

```

```
df1.join(df2,'id').select("id").groupBy("id").agg(count("*")).explain()
```

```
spark_enable_AQE=SparkSession.builder.appName("Data_skewness_before_enabling_AEQ").getOrCreate()

spark_enable_AQE.conf.set("spark.sql.adaptive.enabled", "true")

spark_enable_AQE.conf.set("spark.sql.adaptive.skewJoin.enabled","true")

df3 = spark_enable_AQE.range(1000000000) \
    .withColumn("id", lit("x"))

extravalues = spark_enable_AQE.range(4)\
    .withColumn("id", lit("y"))

moreextravalues = spark_enable_AQE.range(4)\
    .withColumn("id", lit("z"))

df3 = df3.union(extravalues).union(moreextravalues)
df3.display(10)

df4=spark_enable_AQE.range(1000000000) \
    .withColumn("id", lit("x"))
extravalues1 = spark_enable_AQE.range(1)\
    .withColumn("id", lit("y"))

moreextravalues1 = spark_enable_AQE.range(1)\
    .withColumn("id", lit("z"))

df4 = df4.union(extravalues1).union(moreextravalues1)


df3.join(df4,'id').select("id").groupBy("id")\
    .agg(count("*"))\
    .show(100, False)


```

```
df3.join(df4,'id').select("id").groupBy("id")\
    .agg(count("*")).explain()
```
